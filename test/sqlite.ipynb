```python
import sqlite3

conn = sqlite3.connect('bank_database.db')
cursor = conn.cursor()
```

```python
# db 생성
cursor.execute("""
CREATE TABLE IF NOT EXISTS users (
    name TEXT NOT NULL,
    phone TEXT,
    age INTEGER,
    job TEXT,
    marital TEXT,
    education TEXT,
    de_fault TEXT,
    balance INTEGER,
    housing TEXT,
    loan TEXT,
    contact TEXT,
    day INTEGER,
    month TEXT,
    duration INTEGER,
    campaign INTEGER,
    pdays INTEGER,
    previous INTEGER,
    poutcome TEXT,
    deposit TEXT
)
""")
```

```python
import pandas as pd

users_df = pd.read_csv('bank_marketing_user.csv')
users_df = users_df.rename(columns={'이름':'name', '전화번호':'phone', 'default': 'de_fault'})
users_df.head()
```

```python
users_df['phone'] = users_df['phone'].str.replace('-', '')
users_df.head()
```

```python
users_df.values
users = [tuple(v) for v in users_df.values]
users
```

```python
users_df.columns
```

```python
users_df.columns.size
```

```python
cursor.executemany("""
INSERT INTO USERS ('name', 'phone', 'age', 'job', 'marital', 'education', 'de_fault',
       'balance', 'housing', 'loan', 'contact', 'day', 'month', 'duration',
       'campaign', 'pdays', 'previous', 'poutcome', 'deposit')
       VALUES 
       (?,?,?,?,?,?,?,?,?,?, ?,?,?,?,?,?,?,?,?)
""", users)

conn.commit()
```
