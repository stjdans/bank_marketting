```python
import pandas as pd
import numpy as np


marketing_df = pd.read_csv('bank_marketing_data.csv')
sample_df = pd.read_csv('bank_marketing_sample.csv')
phone_df = pd.read_csv('korean_phone_numbers_500.csv')
name_df = pd.read_csv('korean_names_500.csv')

marketing_df.head()
# ['나이', '직업' '결혼여부', '교육수준', '신용불량', '은행잔액', '주택대출', '개인대출', '연락', '일',
# '월', '기간', '캠페인연락횟수', '이전캠페인 연락 받은 횟수', '이전 캠페인 연락 횟수', '이전캠페인 결과', '고객 반등']
```

1. age: 고객의 나이
2. job: 고객의 직업
3. marital: 결혼 여부
4. education: 교육 수준
5. default: 신용 불량 여부
6. balance: 고객의 은행 계좌 잔액
7. housing: 주택 대출 여부
8. loan: 개인 대출 여부
9. contact: 고객과의 연락 방식
10. day: 마지막 연락 일
11. month: 마지막 연락 월
12. duration: 마지막 연락의 지속 시간
13. campaign: 캠페인 기간 동안 고객에게 연락한 횟수
14. pdays: 이전 캠페인에서 고객이 연락을 받은 날 수
15. previous: 이전 캠페인의 연락 횟수
16. poutcome: 이전 캠페인의 결과
17. deposit: 고객의 반응 (예: 예금 계좌 개설 여부)
